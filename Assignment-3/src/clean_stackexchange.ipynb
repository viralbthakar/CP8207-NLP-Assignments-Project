```python
import pandas as pd
import re
```

```python
def stackexchange_clean_CSV(csv_file):
    #CSV file schema: Question ID,Question Title,Tags,Accepted Answer ID,Accepted Answer Body
    csv_df = pd.read_csv(csv_file)
    #Define regex
    # as per recommendation from @freylis, compile once only
    CLEANR = re.compile('<.*?>') 
    csv_df["Question Title"] = csv_df["Question Title"].apply(lambda x: re.sub(CLEANR, '', x))
    csv_df["Accepted Answer Body"] = csv_df["Question Title"].apply(lambda x: re.sub(CLEANR, '', x))

    #Retrieve Query from filename
    tag_searched_for = csv_file.split("_")[1].split(".")[0]
    csv_df["Queried Tag"] = tag_searched_for
    return csv_df
```

```python
got = stackexchange_clean_CSV("../Data/StackExchangeQuestions/SciFi_game-of-thrones.csv")
asoiaf = stackexchange_clean_CSV("../Data/StackExchangeQuestions/SciFi_a-song-of-ice-and-fire.csv")
```
